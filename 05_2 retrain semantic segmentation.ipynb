##Test the model

```
#0006R0_f01170
test_image = "CamVid/test/Seq05VD_f03540.png"
mask_image = "Seq05VD_f03540_pred.png"
output_image = "Seq05VD_f03540_output.png"
```

```
#!python predict.py --image "CamVid/test/Seq05VD_f04980.png" --model "FC-DenseNet56" --checkpoint_path "checkpoints/latest_model_FC-DenseNet56_CamVid.ckpt"
!python predict.py --image "CamVid/test/Seq05VD_f03540.png" --model "DeepLabV3_plus" --checkpoint_path "checkpoints/latest_model_DeepLabV3_plus_CamVid.ckpt"
```

```
#generate the output image
from PIL import Image
img = Image.open(mask_image) #mask
img_o = Image.open(test_image) #original image
width, height = img_o.size 
img = img.resize((width,height)) #resize the mask

px = img.load()
px_o = img_o.load()
#replace the mask part with the mask color
for x in range(0,img_o.size[0]):
    for y in range(0,img_o.size[1]):
        if px[x,y] != (0, 0, 0):
            px_o[x,y] = int((px_o[x,y][0] + px[x,y][0]) / 2), int((px_o[x,y][1] + px[x,y][1]) / 2), int((px_o[x,y][2] + px[x,y][2]) / 2)

img_o.save(output_image)
```

```
#show the images
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 4))

img = Image.open(test_image)
plt.subplot(1, 3, 1)
plt.imshow(img)

img = Image.open(mask_image)
plt.subplot(1, 3, 2)
plt.imshow(img)

img = Image.open(output_image)
plt.subplot(1, 3, 3)
plt.imshow(img)
```
